# Visualisation

```python
df.columns
```

```python
#changing the date column to datetime
df['Date']= pd.to_datetime(df['Date'])
```

```python
# Create Year-Month column
df['Year_Month'] = df['Date'].dt.to_period('M')

# Monthly sales
monthly_sales = df.groupby('Year_Month')['Total Amount'].sum().reset_index()

# Convert Period to Timestamp for plotting
monthly_sales['Year_Month'] = monthly_sales['Year_Month'].astype(str)
```

```python
# Plot
plt.figure(figsize=(12,6))
plt.plot(monthly_sales['Year_Month'],monthly_sales['Total Amount'],marker='o',linewidth=2)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig("images/monthly_sales.png")
plt.show()
```

From the above chart we can see that the highest sales was recorded in May,2023 and the lowest was recorded in January, 2024

```python
# Create Quarter column
df['Quarter'] = df['Date'].dt.to_period('Q')

# Quarterly sales
quarterly_sales = df.groupby('Quarter')['Total Amount'].sum().reset_index()

quarterly_sales['Quarter'] = quarterly_sales['Quarter'].astype(str)
```

```python
# Plot
plt.figure(figsize=(8,5))
plt.plot(quarterly_sales['Quarter'],quarterly_sales['Total Amount'],marker='o',linewidth=2,color='green')

plt.title("Quarterly Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()
plt.savefig("images/quarterly_sales.png")
plt.show()
```

From the above plot we can see that highest sales were recorded in the 4th quarter of the year 2023, and the least in 1st quarter of 2024

```python
#Adding the column age group derived from the age column
bins = [0, 18, 25, 35, 45, 55, 100]
labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '55+']

df['Age Group'] = pd.cut(df['Age'],bins=bins,labels=labels,include_lowest=True)
```

```python
ax= sns.countplot(x= 'Age Group', data= df)

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("customer distribution by age group")
plt.savefig('images/age_grp_distribution.png')
```

Most of the customers are between the age 46-55

```python
sales_age= df.groupby(['Age Group'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False)

sns.barplot(data= sales_age, x='Age Group',y= 'Total Amount', palette='Set2')
plt.savefig('images/age_grp_sales.png')
plt.grid(True)
plt.title("Sales distribution by age group")
plt.show()
```

People belonging to the age group 46-55 have highest contribution in the total sales

```python
#gender wise distribution of people in each age group
ax= sns.countplot(x= 'Age Group', data= df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Customer distribution by age group and gender")
plt.savefig('images/gender_agegrp_distribution.png')
```

```python
sales_gender= df.groupby(['Gender'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False)

sns.barplot(data= sales_gender, x='Gender',y= 'Total Amount', palette='Set2')
plt.savefig('images/gender_sales.png')
plt.grid(True)
plt.title("gender distribution by sales")
plt.show()
```

Females have the highest contribution in the total sales

```python
df['Product Category'].unique()
```

```python
sales_prod=df.groupby(['Product Category'], as_index=False)['Total Amount'].sum()

sns.barplot(data= sales_prod, x='Product Category',y='Total Amount', palette='Set2')
plt.savefig('images/Product_category_sales.png')
plt.title("Prodcut wise distribution of sales")
plt.show()
```

The dataset contains only three product categories (Beauty, Clothing, and Electronics) and does not include individual product names or IDs. Therefore, a "Top 10 Best-Selling Products" analysis is not possible. Instead, the analysis focuses on the best-selling product categories.

We can see that Electronics category has contributed highest in the Total revenue

```python
best_selling = df.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False)
best_selling.plot(kind='bar')

plt.title('Best-Selling Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=0)
plt.savefig('images/quantity_wise_sales.png')
plt.show()
```

Most of the sold products are from clothing category

```python
# Select numerical columns
numerical_cols = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']
corr_matrix = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)

plt.title('Correlation Heatmap of Numerical Features')
plt.savefig('images/correlation_heatmap.png')
plt.show()
```

We can see that:
- Total Amount has a strong positive correlation with Quantity and Price per Unit, since it is calculated from these variables.
- Age generally shows a weak or negligible correlation with sales-related variables, indicating that customer age does not strongly influence purchase quantity or transaction value
- Quantity and Price per Unit may exhibit a weak or near-zero correlation, suggesting that higher-priced products are not necessarily purchased in larger quantities

```python
monthly_aov = df.groupby(df['Date'].dt.to_period('M'))['Total Amount'].mean().reset_index()

monthly_aov['Date'] = monthly_aov['Date'].astype(str)

plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_aov, x='Date', y='Total Amount', marker='o')

plt.title('Average Order Value by Month')
plt.xlabel('Month')
plt.ylabel('Average Order Value')
plt.xticks(rotation=45)
plt.savefig('images/aov_month.png')
plt.show()
```

Sales growth may be driven by customers spending more per purchase rather than an increase in the number of transactions.
